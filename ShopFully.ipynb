#2) Estrazione dati

```python
# https://api.openweathermap.org/data/3.0/onecall/timemachine?lat={lat}&lon={lon}&dt={time}&appid={API key}
open_w = "https://api.openweathermap.org/data/3.0/onecall/"
#timemachine?lat={lat}&lon={lon}&dt={time}&appid={API key}
def get_data(city, timestamp):
  url = f"{open_w}timemachine?lat={city['lat']}&lon={city['lon']}&dt={timestamp}&appid={appid}"
  data = requests.get(url)
  return data.json()
```

```python
year = 2024
month = 10

timestamps = []

for day in range(1, 32):
    for hour in range(24):
        timestamp = dt.datetime(year, month, day, hour).timestamp()
        timestamps.append(timestamp)
```

```python
len(timestamps)
```

```python
data_Bologna = []
data_Milano = []
data_Cagliari = []
# Print the last_days
for ts in timestamps:
  result = get_data(cities['Bologna'], int(ts))
  result = get_data(cities['Milano'], int(ts))
  result = get_data(cities['Cagliari'], int(ts))
  data_Bologna.append(result['data'][0])
  data_Milano.append(result['data'][0])
  data_Cagliari.append(result['data'][0])

```

```python
df_bologna = pd.DataFrame(data=data_Bologna)
df_milano = pd.DataFrame(data=data_Milano)
df_cagliari = pd.DataFrame(data=data_Cagliari)
```

```python
data_bologna_weather = []
data_milano_weather = []
data_cagliari_weather = []
for row in data_Bologna:
  for i in row['weather']:
    data_bologna_weather.append(i)
for row in data_Milano:
  for i in row['weather']:
    data_milano_weather.append(i)
for row in data_Cagliari:
  for i in row['weather']:
    data_cagliari_weather.append(i)

df_bologna_weather = pd.DataFrame(data_bologna_weather).drop_duplicates()
df_milano_weather = pd.DataFrame(data_bologna_weather).drop_duplicates()
df_cagliari_weather = pd.DataFrame(data_bologna_weather).drop_duplicates()
```

##4.5)  Which city had the highest daily temperature variation in a certain period of time?

```python
start_date = '2024-10-15'
# Filter the DataFrame for the specified date range
df_filtered = df_merged[(df_merged['dt'] >= start_date)]

# Extract date from datetime for grouping
df_filtered.loc[:,'date'] = df_filtered['dt'].dt.date
df_filtered
```

```python
daily_temps = df_filtered.groupby(['city_id', 'date']).agg(max_temp=('temp', 'max'), min_temp=('temp', 'min')).reset_index()


daily_temps['temp_variation'] = daily_temps['max_temp'] - daily_temps['min_temp']


max_variation_city = daily_temps.loc[daily_temps['temp_variation'].idxmax()]
```

```python
print(f"City ID: {max_variation_city['city_id']}, "
      f"Max Temperature Variation: {max_variation_city['temp_variation']}")
```
